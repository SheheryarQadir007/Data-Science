```python
data['InvoiceDate'] = data['Description'].fillna('Unknown')
data = data.dropna(subset=['CustomerID'])
data.isnull().sum()
```

```python
data = data.drop_duplicates()
data.head()
```

```python
data['TotalAmount'] = data['Quantity'] * data['UnitPrice']
data.head()
```

```python
customer_data = data.groupby('CustomerID').agg({
    'TotalAmount': 'sum',
    'InvoiceNo': 'nunique',
    'Quantity': 'mean'
}).reset_index()
customer_data.columns = ['CustomerID', 'TotalSpending', 'NumTransactions', 'AvgBasketSize']
customer_data.head()
```

```python
from sklearn.preprocessing import StandardScaler

features = ['TotalSpending', 'NumTransactions', 'AvgBasketSize']
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[features])
print(scaled_data)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia=[]
k = range(1, 11)
for i in k:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k, inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(k)
plt.grid()
plt.show()
```

```python
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(scaled_data)
customer_data.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

plt.figure(figsize=(10, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=customer_data['Cluster'], cmap='viridis', marker='o')
plt.title('Customer Segmentation')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
cluster_analysis = customer_data.groupby('Cluster').mean()
print(cluster_analysis)
```

```python
import seaborn as sns
cluster_analysis_melted = cluster_analysis.reset_index().melt(id_vars='Cluster')

plt.figure(figsize=(12, 6))
sns.barplot(x='Cluster', y='value', hue='variable', data=cluster_analysis_melted, palette='viridis')
plt.title('Average Cluster Analysis')
plt.xlabel('Cluster')
plt.ylabel('Average Value')
plt.legend(title='Features')
plt.grid()
plt.show()
```
